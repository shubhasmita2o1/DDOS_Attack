```python
from catboost import CatBoostClassifier 
import pandas as pd
from sklearn.model_selection import train_test_split
import pickle
```

```python
#load the traffic file
trafficData = pd.read_csv('dataset/traffic.csv')
trafficData =  trafficData.drop(['Unnamed: 0'], axis=1)
trafficData.head()

labelOut = list(set(trafficData['label'].values.tolist()))
labelOutIndex =  range(0,len(labelOut))
labelOutInfo = dict(zip(labelOut,labelOutIndex))
trafficData['label'] =  trafficData['label'].map(labelOutInfo)

fname = 'Category/traffic_Category_data.sav'
pickle.dump(labelOutInfo, open(fname, 'wb'))
```

```python

predict_label = trafficData.label
X = trafficData.drop('label', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, predict_label ,test_size=0.2)

print(X_test.values[1])
print( y_test.values[1])

filename = 'traffic_test_Data.sav'
pickle.dump(X_test, open(filename, 'wb'))
filename1 = 'traffic_test_label.sav'
pickle.dump(y_test, open(filename1, 'wb'))
```

```python
X_test
```

```python
model_CBC = CatBoostClassifier()
model_CBC.fit(X_train, y_train)
```

```python
filename = 'Model/catboost_traffic.sav'
pickle.dump(model_CBC, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open('Model/catboost_traffic.sav', 'rb'))
traffic_test =  pickle.load(open('random/traffic_test_Data.sav', 'rb'))
traffic_test_label =  pickle.load(open('random/traffic_test_label.sav', 'rb'))
```

```python
import numpy as np
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score
# predict on test data

y_pred = loaded_model.predict(X_test)
acc = accuracy_score(y_test, y_pred)



macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(y_test, y_pred, average='macro')



print("Accuracy:{:.6f}, Precision: {:.4f}, Recall: {:.4f}, F1-Score: {:.4f}".format(acc,macro_precision, macro_recall, macro_f1))
```

```python
performanceMetrics =  acc, macro_precision, macro_recall, macro_f1
filename = 'PerformanceMetric/catboost_traffic.sav'
pickle.dump(performanceMetrics, open(filename, 'wb'))
```
