```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pickle
#load the traffic file
ddosData = pd.read_csv('dataset/ddosattack_malicious.csv')
ddosData =  ddosData.drop(['Unnamed: 0'], axis=1)
labelOut = list(set(ddosData['label'].values.tolist()))
labelOutIndex =  range(0,len(labelOut))
labelOutInfo = dict(zip(labelOut,labelOutIndex))
ddosData['label'] =  ddosData['label'].map(labelOutInfo)
ddosData.head()
```

```python
def datasetParameters(trafficData):
    predict_label = trafficData.label
    X = trafficData.drop('label', axis=1)
    X_train, X_test, y_train, y_test = train_test_split(X, predict_label ,test_size=0.2)
    data = X_train, X_test, y_train, y_test
    return data
```

```python
def generateRandomForestModel(X_train, y_train,filename):
    RF = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 42)  
    RF.fit(X_train, y_train)
    filename = 'Model/'+filename+'.sav'
    pickle.dump(RF, open(filename, 'wb'))
    return RF
```

```python
X_train, X_test, y_train, y_test = datasetParameters(ddosData)
VV = generateRandomForestModel(X_train, y_train,'ddos_rf_model')
print('Model Created Successfully for ddos dataset using Random Forest')
```

```python
loaded_model = pickle.load(open('Model/ddos_rf_model.sav', 'rb'))
```

```python
filename = 'random/ddos_test_Data.sav'
pickle.dump(X_test, open(filename, 'wb'))
filename1 = 'random/ddos_test_label.sav'
pickle.dump(y_test, open(filename1, 'wb'))
```

```python
ddos_test =  pickle.load(open('random/ddos_test_Data.sav', 'rb'))
ddos_label =  pickle.load(open('random/ddos_test_label.sav', 'rb'))
```

```python
import numpy as np
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score
# predict on test data
y_pred = loaded_model.predict(ddos_test)
acc = accuracy_score(ddos_label, y_pred)


# macro average
macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(y_test, y_pred, average='macro')

# Print the results

print("Accuracy:{:.6f}, Precision: {:.4f}, Recall: {:.4f}, F1-Score: {:.4f}".format(acc,macro_precision, macro_recall, macro_f1))
```
