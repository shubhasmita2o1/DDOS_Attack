```python
# splitting datasets into training and testing
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pickle
#load the traffic file
trafficData = pd.read_csv('dataset/traffic.csv')
trafficData =  trafficData.drop(['Unnamed: 0'], axis=1)
labelOut = list(set(trafficData['label'].values.tolist()))
labelOutIndex =  range(0,len(labelOut))
labelOutInfo = dict(zip(labelOut,labelOutIndex))
trafficData['label'] =  trafficData['label'].map(labelOutInfo)
trafficData.head()
```

```python
def generateRandomForestModel(X_train, y_train,filename):
    RF = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 42)  
    RF.fit(X_train, y_train)
    filename = 'Model/'+filename+'.sav'
    pickle.dump(RF, open(filename, 'wb'))
    return RF
```

```python

def datasetParameters(Data):
    predict_label = Data.label
    X = Data.drop('label', axis=1)
    X_train, X_test, y_train, y_test = train_test_split(X, predict_label ,test_size=0.2)
    totalData = X_train, X_test, y_train, y_test
    return totalData
```

```python
X_train, X_test, y_train, y_test = datasetParameters(trafficData)

VV = generateRandomForestModel(X_train, y_train,'traffic_rf_model')
print('Model Created Successfully for Traffic dataset using Random Forest')
```

```python
filename = 'random/traffic_test_Data.sav'
pickle.dump(X_test, open(filename, 'wb'))
filename1 = 'random/traffic_test_label.sav'
pickle.dump(y_test, open(filename1, 'wb'))
```

```python
loaded_model = pickle.load(open('Model/traffic_rf_model.sav', 'rb'))
traffic_test =  pickle.load(open('random/traffic_test_Data.sav', 'rb'))
traffic_test_label =  pickle.load(open('random/traffic_test_label.sav', 'rb'))
```

```python
import numpy as np
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score
# predict on test data
y_pred = loaded_model.predict(traffic_test)
acc = accuracy_score(y_test, traffic_test_label)

macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(y_test, y_pred, average='macro')

# Print the results

print("Accuracy:{:.6f}, Precision: {:.4f}, Recall: {:.4f}, F1-Score: {:.4f}".format(acc,macro_precision, macro_recall, macro_f1))
```
